```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Question 1

def VectorLength(v):
    squares = np.power(v,2)
    sum_square = np.sum(squares)
    result = np.sqrt(sum_square)
    
    return result 

v_1 = (5,6)
Length_v_1 = VectorLength(v_1)
print(Length_v_1)

v_2 = (-1,5)
Length_v_2 = VectorLength(v_2)
print(Length_v_2)


sin_one = np.sin(1)
print("Sin(1) = ", np.sin(1))
print(np.round_(sin_one, decimals = 3))
```

## Question 2


```python
"""
The IsPrime checks a if n is a prime number. 
It returns 0 if n is composite and 1 is n is prime.

I fixed this function by changing the '<' which appears after the while loop to a '<='.
This is because to check prime numbers, divisors of up to and including square root of
n need to be checked.

For example for n = 9 , the old function would compute 3<3 and return 1 instead of checking if 3 divides 3.
I also moved the last return statement outside of the loop.
"""
def IsPrime(n):
    if (n == 1):
        return 0
    i = 2
    while i <= np.sqrt(n):
        # Check each i check if it divides n. 
        if(n % i == 0):
            return 0 
        i += 1
        # If no divisors are found, the number is prime. 
    return 1

# Part a - Checking if function work from 1-10.
for i in range(1,11):
    print(IsPrime(i))
    
# Part e - Iotal prime numbers less than 1000.
total_count = 0

for i in range(1,1000):
        if IsPrime(i):
            total_count += 1
            
print('There are', total_count,'prime numbers less than 1000')

# Part f - Function with number of primes less than or equal to x.
def PrimeCount(x):
    total_count = 0
    for i in range(1,x):
        if IsPrime(i):
            total_count += 1
    return total_count 

PrimeCount(1000)
PrimeCount(10000)

# Part g - Plot the counting function.
x = np.arange(0,40,1)
Pi = np.zeros(40)
i = 0
while i < np.size(Pi):
    Pi[i] = PrimeCount(i)
    i += 1

plt.plot(x, Pi, 'ro')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('Pi(x)')


 
```

```python
# Part b - Loading file and defining x and y variables.
fname = 'PowerLawOrExponential (1).txt'
a = np.loadtxt(fname)
a

type(a)

# Using transpose method to get x and y variable.
x = a.T[0]
y = a.T[1]

# Part c - Graphing data.
log_y = np.log10(y)
log_x = np.log10(x)

plt.plot(x, log_y)
plt.grid(True)
plt.title('X Vs log(y)')
plt.xlabel('X')
plt.ylabel('Log(y)')
plt.show()

plt.plot(log_x, log_x)
plt.grid(True)
plt.title('Log(x) Vs Log(y)')
plt.xlabel('Log(x)')
plt.ylabel('Log(y)')
plt.show()

```
